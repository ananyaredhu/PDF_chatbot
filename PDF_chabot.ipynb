```python
import urllib
import warnings
from pathlib import Path as p
from pprint import pprint

import pandas as pd
from langchain import PromptTemplate
from langchain.chains.question_answering import load_qa_chain
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.chains import RetrievalQA

warnings.filterwarnings("ignore") #to restart python kernal if issues occur with langchain import

```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
model = ChatGoogleGenerativeAI(model = "gemini-pro", google_api_key = GOOGLE_API_KEY,
                               temperature = 0.7, convert_system_message_to_human=True)
```

```python
import sqlite3
```

```python
# Connect to the SQL database
conn = sqlite3.connect("faq_database.db")
cursor = conn.cursor()

# Create the FAQ table if it doesn't exist
cursor.execute("CREATE TABLE IF NOT EXISTS faq (question TEXT, answer TEXT)")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=1000)  # Adjust chunk sizes as needed
```

```python
# Function to load and index multiple PDFs

def load_and_index_pdfs(pdf_paths):
    pages = []
    texts = []
    for pdf_path in pdf_paths:
        pdf_loader = PyPDFLoader(pdf_path)
        new_pages = pdf_loader.load_and_split()
        pages.extend(new_pages)
        context = "\n\n".join(str(p.page_content) for p in new_pages)
        texts.extend(text_splitter.split_text(context))

    vector_index = Chroma.from_texts(texts, embeddings).as_retriever(search_kwargs={"k": 5})
    return vector_index
```

```python
pdf_paths = ["/content/attention.pdf", "/content/TextureDreamer.pdf", "/content/Open-Vocab.pdf", "/content/vlogger.pdf"]  # Add more PDF paths here
vector_index = load_and_index_pdfs(pdf_paths)
```

```python
embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001",google_api_key=GOOGLE_API_KEY)
```

RAG pipeline: Embedding + Gemini(LLM)

```python
# Function to add a new FAQ to the database
def add_faq(question, answer):
  cursor.execute("INSERT INTO faq (question, answer) VALUES (?, ?)", (question, answer))
  conn.commit()
```

```python
# Function to retrieve an answer from the database
def get_answer_from_database(question):
  cursor.execute("SELECT answer FROM faq WHERE question=?", (question,))
  result = cursor.fetchone()
  if result:
    return result[0] # Return the answer if found
  else:
    return None
```

```python
template = """Use the following pieces of context to provide a detailed and comprehensive answer to the question at the end. Be as informative as possible, including relevant examples and key points. If you need more information to provide a thorough answer, please state what is missing. Always say "thanks for asking!" at the end of the answer.
Context: {context}
Question: {question}
Helpful Answer:"""
QA_CHAIN_PROMPT = PromptTemplate.from_template(template)# Run chain
qa_chain = RetrievalQA.from_chain_type(
  model,
  retriever=vector_index,
  return_source_documents=True,
  chain_type_kwargs={"prompt": QA_CHAIN_PROMPT}
)
```

```python
# Main loop for answering questions
while True:
    question = input("Ask a question (or type 'exit' to quit): ")
    if question.lower() == "exit":
        break

    answer = get_answer_from_database(question)
    if answer:
        print("Answer from database:", answer)
    else:
        try:
            result = qa_chain({"query": question})  # Call the RAG pipeline
            print("Generated answer from Gemini API:", result["result"])


            # Add the generated answer to the database
            try:
                add_faq(question, result["result"])
                print("Answer added to database for future use.")
            except sqlite3.Error as e:
                print(f"Error adding answer to database: {e}")

        except Exception as e:
            print(f"Error using Gemini API: {e}")
```
